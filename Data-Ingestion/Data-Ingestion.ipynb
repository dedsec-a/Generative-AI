```python
# Trying the WebPage Loader 
from langchain_community.document_loaders import WebBaseLoader
import bs4
loader = WebBaseLoader(web_path=("https://en.wikipedia.org/wiki/Generative_artificial_intelligence"),bs_kwargs=dict(parse_only = bs4.SoupStrainer(
    class_ = ("mw-page-container")
)))
```

```python
docs = loader.load()
```

```python
docs
```
